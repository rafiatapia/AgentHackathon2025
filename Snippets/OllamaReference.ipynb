#### Usefull Commands

- **List Models**
```
ollama list
```

- **Remove Models**
```
ollama rm llama3
```

- **Pull a model**
```
ollama pull llama3.1
```


- **Run Model To access via Console**
```
ollama run llama3
```

- **Run Ollama in the background**
```
ollama serve
```

- **Query the model through web api**
```
curl http://localhost:11434/api/generate -d '{"model": "llama3.2:1b","prompt":"What is the capital of Pakistan"}'
```

- **Using chat and rest api endpoint**
```
curl http://localhost:11434/api/chat -d '{"model": "llama3.2:1b","prompt":"What is the capital of Pakistan"}'
```
